```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, export_text
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.impute import SimpleImputer

import matplotlib.pyplot as plt
```

Dataset

```python
data=pd.read_csv('/content/vote.csv')
data.head()
```

```python
import numpy as np
transactions =[]
for i in range(0,435):
  transactions.append([str(data.values[i,j]) for j in range(0,17)])

print(transactions)
```

```python
from mlxtend.preprocessing import TransactionEncoder
te= TransactionEncoder()
te_arry = te.fit(transactions).transform(transactions)
df= pd.DataFrame(te_arry, columns=te.columns_)
df
```

```python
from mlxtend.frequent_patterns import apriori
ruleset = apriori(df, min_support=0.0045, use_colnames=True)
ruleset["length"]= ruleset["itemsets"].apply(lambda x: len(x))
ruleset
```

```python
ruleset[(ruleset["length"]==3)&(ruleset["support"]>0.05)]
```

```python
ruleset[ruleset["itemsets"] == frozenset({"democrat", "y"})]
```

```python
from mlxtend.frequent_patterns import association_rules
rules = association_rules(ruleset, metric="lift", min_threshold=1)
rules= rules.sort_values(["confidence","lift"], ascending=[False,False])
display(rules)
```
