```python
# @title AI prompt cell

import ipywidgets as widgets
from IPython.display import display, HTML, Markdown,clear_output
from google.colab import ai

dropdown = widgets.Dropdown(
    options=[],
    layout={'width': 'auto'}
)

def update_model_list(new_options):
    dropdown.options = new_options
update_model_list(ai.list_models())

text_input = widgets.Textarea(
    placeholder='Ask me anything....',
    layout={'width': 'auto', 'height': '100px'},
)

button = widgets.Button(
    description='Submit Text',
    disabled=False,
    tooltip='Click to submit the text',
    icon='check'
)

output_area = widgets.Output(
     layout={'width': 'auto', 'max_height': '300px','overflow_y': 'scroll'}
)

def on_button_clicked(b):
    with output_area:
        output_area.clear_output(wait=False)
        accumulated_content = ""
        for new_chunk in ai.generate_text(prompt=text_input.value, model_name=dropdown.value, stream=True):
            if new_chunk is None:
                continue
            accumulated_content += new_chunk
            clear_output(wait=True)
            display(Markdown(accumulated_content))

button.on_click(on_button_clicked)
vbox = widgets.GridBox([dropdown, text_input, button, output_area])

display(HTML("""
<style>
.widget-dropdown select {
    font-size: 18px;
    font-family: "Arial", sans-serif;
}
.widget-textarea textarea {
    font-size: 18px;
    font-family: "Arial", sans-serif;
}
</style>
"""))
display(vbox)
```

```python
import datetime

def get_planting_advice(province, month=None):
    if month is None:
        month = datetime.datetime.now().month

    # Define South African Seasons
    seasons = {
        "Summer": [12, 1, 2],
        "Autumn": [3, 4, 5],
        "Winter": [6, 7, 8],
        "Spring": [9, 10, 11]
    }

    # Determine current season
    current_season = next(s for s, m in seasons.items() if month in m)

    # Crop Database based on SA Climate Zones
    crops = {
        "Gauteng": {
            "Summer": "Swiss Chard, Tomatoes, and Peppers (Watch for afternoon hailstorms!)",
            "Winter": "Cabbage, Peas, and Broccoli (Protect from morning frost.)",
            "Spring": "Carrots, Green Beans, and Maize.",
            "Autumn": "Beetroot, Lettuce, and Onions."
        },
        "Western Cape": {
            "Summer": "Eggplants, Chillies, and Squash (Water early to avoid evaporation).",
            "Winter": "Leeks, Garlic, and Broad Beans (Plant in well-drained soil for rain).",
            "Spring": "Cucumbers, Potatoes, and Herbs.",
            "Autumn": "Spinach, Kale, and Turnips."
        },
        "KwaZulu-Natal": {
            "Summer": "Sweet Potatoes, Okra, and Maize (High humidity favorites).",
            "Winter": "Almost anything! Try Peppers and Tomatoes in the mild winter.",
            "Spring": "Pumpkin, Beans, and Melons.",
            "Autumn": "Cabbage, Cauliflower, and Carrots."
        }
    }

    # Default logic for provinces not explicitly listed
    zone_advice = crops.get(province, crops["Gauteng"])
    advice = zone_advice.get(current_season)

    return f"In {province}, it is currently {current_season}. You should plant: {advice}"

# Example Usage:
user_province = "Western Cape"
print(get_planting_advice(user_province))
```

```python
import os

import markdown
import faiss
import torch
from bs4 import BeautifulSoup
from sentence_transformers import SentenceTransformer
```

##### Downloading your documents

##### Splitting your documents

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
def md_to_text(md):
    html = markdown.markdown(md)
    soup = BeautifulSoup(html, features='html.parser')
    return soup.get_text()

def split_text(document_path, chunk_size, overlap):
  with open(document_path, 'r', encoding='utf-8') as file:
        content = file.read()

  content = md_to_text(content)
  chunks = []
  for i in range(0, len(content), chunk_size):
    chunk = document_path + content[max(0, i-overlap):min(len(content), i+chunk_size)] #pymatting.md content[0:100]. -> pymatting.md content[100:200] ...
    chunks.append(chunk)
  return chunks
```

```python
readmes_dir = "readmes"
chunk_size = 1000
overlap = 200

all_chunks = []
for f in os.listdir(readmes_dir):
  if f[-2:] == "md":
    all_chunks += split_text(os.path.join(readmes_dir, f), chunk_size, overlap)
```

```python
len(all_chunks)
```

##### Indexing your documents

```python
embedding_model = SentenceTransformer("all-MiniLM-L6-v2")
_ = embedding_model.to(device)
embeddings = embedding_model.encode(all_chunks)
embeddings.shape
```

```python
d = embeddings.shape[1] # 384
index = faiss.IndexFlatL2(d)
index.add(embeddings)
```

##### Building your RAG prompt

```python
prompt = """Who were the authors of pymatting library?"""
embedding = embedding_model.encode([prompt])
_, I = index.search(embedding, 5)

context = "Relevant documents:\n"
for i in I[0]:
  context += f"Doc {i+1}: {all_chunks[i]}\n"

final_prompt = f"Use the documents to answer the user.\n{context}\n{prompt}"

for chunk in client.models.generate_content_stream(
    model = model,
    contents = final_prompt,
    config = generate_content_config,
    ):
    print(chunk.text, end="")
```

## 1. Core Features of the Agribusiness Platform

This section outlines the essential functionalities required for an agribusiness app connecting farmers across Africa, based on the initial problem statement and objective.

### A. Farmer Profiles and Networking

*   **User Registration & Profiles**: Secure onboarding for farmers, agribusinesses, and other stakeholders. Profiles should include location, crops/products, contact information, and specific needs/offers.
*   **Search & Discovery**: Functionality to find other farmers, buyers, suppliers, and agricultural experts based on location, crop type, demand/supply, etc.
*   **Direct Messaging/Chat**: In-app communication to facilitate direct interaction between users.

### B. Marketplace Functionality

*   **Product Listings**: Farmers can list their produce, livestock, or agricultural products with details like quantity, price, availability, and quality.
*   **Request for Products/Services**: Buyers and agribusinesses can post specific needs or tenders for agricultural goods and services.
*   **Price Discovery & Market Information**: Provide real-time or near real-time market prices for various commodities to empower farmers in negotiation.
*   **Logistics & Delivery (Optional/Integration)**: Potential integration with logistics providers or a basic system for farmers to indicate delivery capabilities/needs.

### C. AI-Powered Advisory Services (Addressing the 'Extension Gap')

*   **Crop Stress Diagnosis**: An AI model (e.g., using image recognition) to help farmers identify pests, diseases, or nutrient deficiencies from photos of their crops.
*   **Personalized Planting Schedules**: AI-driven recommendations for optimal planting times, crop rotation, and suitable varieties based on local climate data, soil type, and historical yield data.
*   **Multilingual Support**: Ensure the AI and app interface support multiple local languages to maximize accessibility.
*   **Offline Functionality**: Critical for rural areas with limited internet connectivity; allowing farmers to use key features and store data offline, syncing when connectivity is available.

### D. Knowledge Sharing & Community

*   **Knowledge Base/Resources**: Articles, guides, and best practices on sustainable farming, pest control, financial management, etc.
*   **Community Forums/Groups**: Spaces for farmers to share experiences, ask questions, and learn from peers and experts.
*   **Weather Forecasts**: Localized weather information relevant to farming activities.

### E. Data & Analytics

*   **Farm Management Tools**: Simple tools for tracking expenses, yields, and farm activities.
*   **Reporting & Analytics**: Basic dashboards for farmers to monitor their sales, market trends, and farm performance.

### F. Technical Considerations

*   **Mobile-First Design**: Given the target audience, the app must be primarily designed for mobile devices (smartphones, potentially feature phones).
*   **Scalability**: The platform should be designed to scale to accommodate a large number of users and transactions across different regions.
*   **Security & Privacy**: Robust measures to protect user data and financial transactions.

### G. User Interface and Experience (UI/UX)

*   **Agriculture-Centric Design**: The interface should be intuitive and designed with the specific needs and context of African farmers in mind, using relevant agricultural imagery, terminology, and workflows. This includes easy navigation, clear visual cues, and a focus on essential functions.
*   **Accessibility**: Design for users with varying levels of digital literacy and access to technology, including considerations for screen size, touch targets, and minimal data usage.
*   **Multilingual Interface**: Beyond AI advisory services, the entire app interface should support multiple local languages to enhance usability and inclusivity.
