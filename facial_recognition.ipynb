#Libraries

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from PIL import Image
import cv2
```

Path Extraction

```python
import zipfile

zip_path = "/content/Datasets.zip"
extract_path = "/content/Datasets"

if not os.path.exists(extract_path):
    os.makedirs(extract_path)

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print(f"Dataset extracted to: {extract_path}")
```

```python
IMAGE_EXTS = ('.jpg', '.jpeg', '.png', '.bmp')


def is_image_file(fname):
    return fname.lower().endswith(IMAGE_EXTS)

IMAGE_SIZE = (64, 64) # Define IMAGE_SIZE here

def load_dataset(dataset_dir, image_size=IMAGE_SIZE, verbose=True):
    X = [] # flattened images
    y = [] # labels (string)
    label_names = []


    for label in sorted(os.listdir(dataset_dir)):
        label_path = os.path.join(dataset_dir, label)
        if not os.path.isdir(label_path):
            continue
        if verbose:
            print(f"Reading label: {label}")
        for fname in os.listdir(label_path):
            if not is_image_file(fname):
                continue
            fpath = os.path.join(label_path, fname)
            img = cv2.imread(fpath)
            if img is None:
                print(f"Warning: couldn't read {fpath}")
                continue
            # convert to grayscale
            gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            # resize
            resized = cv2.resize(gray, image_size)
            # flatten
            flat = resized.flatten().astype('float32') / 255.0
            X.append(flat)
            y.append(label)
        label_names.append(label)


    X = np.array(X)
    y = np.array(y)
    if verbose:
        print(f"Loaded {len(X)} images, labels: {np.unique(y)}")
    return X, y
```
